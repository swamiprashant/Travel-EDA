```python
cats = []
nums = []
for i in df.columns:
    
    if df[i].dtypes == 'object':
        cats.append(i)
    else:
        nums.append(i)
```

```python
for cat_column in cats:
    plt.figure(figsize=(8, 6))
    sns.countplot(x= cat_column, data = df, palette = 'viridis')
    plt.title(f'Univariate analysis- {cat_column}')
    plt.show()
```

```python
# Insight: Salaried and small business people travale more than Large business and freelance. both male and female are mostly like to travel deluxe and basic coach. married are mostly travel.
```

```python
nums
```

```python
for num_column in nums:
    plt.figure(figsize = (8, 6))
    sns.histplot(df[num_column], kde = True, color = 'skyblue')
    plt.title(f'univariate analysis - {num_column}')
    plt.show()
```

```python
# Insight : monthly income of mejority of travelers are between 20k-40k.
#Age of mostly travelers are between 30-50 years.
#number of folloups are between 3-5.
#Number of trips per month is between 1-8.
```
